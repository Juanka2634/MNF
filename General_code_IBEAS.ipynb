```python
# Create BIDS with IBEAS images
# Unzip files
import os
import subprocess
from os import listdir
from os.path import isfile, join
```

```python
subjects = []
img_whitout_raw = []
subj_path = '/opt/dora/Dora/IBEAS/IBEAS_Cloud'
subjects = os.listdir(subj_path)
subjects.sort()
print(len(subjects))
```

```python
for x in subjects[2:]:
    subj_path_img = subj_path+'/'+x
    subprocess.run(['mkdir', subj_path_img+'/raw/'])
    img_whitout_raw = [f for f in listdir(subj_path_img) if isfile(join(subj_path_img,f))]
    for y in img_whitout_raw:
        #print(subj_path_img+'/raw/'+y+'.dcm')
        #print(subj_path_img+y)
        subprocess.run(['gdcmconv','--raw', subj_path_img+'/'+y, subj_path_img+'/raw/'+y+'.dcm'])
        subprocess.run(['rm', subj_path_img+'/'+y])
        subprocess.run(['cp', subj_path_img+'/raw/'+y+'.dcm', subj_path_img+'/'+y+'.dcm'])
    #print(img_whitout_raw)
```
