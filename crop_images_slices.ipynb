```python
from PIL import Image
import os.path, sys
```

```python
path = "/opt/dora/Dora/IBEAS/Resultados/img_results_presentation/Asl/"
# image_secuence = "DTI"
dirs_image_1 = path+"controles-pacientes_asl_cluster_correct(11).png"
dirs_image_2 = path+"controles-pacientes_asl_cluster_correct(12).png"
dirs_image_3 = path+"controles-pacientes_asl_cluster_correct(13).png"
file_output = path +'controles-pacientes_asl_cluster_correct(collage1).bmp'
```

```python
#Normal Images
# Process image_1---------------------------------------------------------
im_1 = Image.open(dirs_image_1)
file_1, ext_1 = os.path.splitext(dirs_image_1)
width, height = im_1.size
#print(width)
#print(height)
# Image_1
left = 1
right = (width/2)-15
top = 1
bottom = height-30
imCrop = im_1.crop((left, top, right, bottom))
imCrop.save(file_1 + 'Cropped1.bmp', "BMP", quality=100)
# Image_2
left = (width/2)-15
right = width
top = 1
bottom = height-10
imCrop = im_1.crop((left, top, right, bottom))
imCrop.save(file_1 + 'Cropped2.bmp', "BMP", quality=100)
# Process image_2---------------------------------------------------------
im_2 = Image.open(dirs_image_2)
file_2, ext_2 = os.path.splitext(dirs_image_2)
width, height = im_2.size
#print(width)
#print(height)
# Image_1
left = 1
right = (width/2)+30
top = 1
bottom = height-20
imCrop = im_2.crop((left, top, right, bottom))
imCrop.save(file_2 + 'Cropped1.bmp', "BMP", quality=100)
# Image_2
left = (width/2)+32
right = width
top = 1
bottom = height-25
imCrop = im_2.crop((left, top, right, bottom))
imCrop.save(file_2 + 'Cropped2.bmp', "BMP", quality=100)
# Process image_3---------------------------------------------------------
im_3 = Image.open(dirs_image_3)
file_3, ext_3= os.path.splitext(dirs_image_3)
width, height = im_3.size
#print(width)
#print(height)
# Image_1
left = 1
right = (width)-115
top = 1
bottom = height-35
imCrop = im_3.crop((left, top, right, bottom))
imCrop.save(file_3 + 'Cropped1.bmp', "BMP", quality=100)
#---------------------------------------------------------
image_crop_1_1 = Image.open(file_1 + 'Cropped1.bmp')
image_crop_1_2 = Image.open(file_1 + 'Cropped2.bmp')
image_crop_2_1 = Image.open(file_2 + 'Cropped1.bmp')
image_crop_2_2 = Image.open(file_2 + 'Cropped2.bmp')
image_crop_3_1 = Image.open(file_3 + 'Cropped1.bmp')
#---------------------------------------------------------
posx_img_1_1 = int(image_crop_1_1.size[0]-(2*image_crop_1_1.size[0]/2)/2)
posy_img_1_1 = int(4*image_crop_1_1.size[1])

posx_img_2_1 = int(image_crop_1_1.size[0]-(2*image_crop_1_2.size[0]/2)/2)
posy_img_2_1 = int(3*image_crop_1_2.size[1])

posx_img_3_1 = int(image_crop_1_1.size[0]-(2*image_crop_2_1.size[0]/2)/2)
posy_img_3_1 = int(2*image_crop_2_1.size[1])

posx_img_4_1 = int(image_crop_1_1.size[0]-(2*image_crop_2_2.size[0]/2)/2)
posy_img_4_1 = int(1*image_crop_2_2.size[1])

posx_img_5_1 = int(image_crop_1_1.size[0]-(2*image_crop_3_1.size[0]/2)/2)
posy_img_5_1 = int(0)
collage = Image.new('RGB',(2*image_crop_1_1.size[0], 5*image_crop_1_1.size[1]+20), (0,0,0))
collage.paste(image_crop_1_1,(posx_img_1_1,posy_img_1_1+20))
collage.paste(image_crop_1_2,(posx_img_2_1,posy_img_2_1-20))
collage.paste(image_crop_2_1,(posx_img_3_1,posy_img_3_1-20))
collage.paste(image_crop_2_2,(posx_img_4_1,posy_img_4_1-10))
collage.paste(image_crop_3_1,(posx_img_5_1,posy_img_5_1))
collage.save(path+"merged_image.bmp","BMP")
left = image_crop_1_1.size[0]-(2*image_crop_2_1.size[0]/2)/2-10
right = image_crop_1_1.size[0]+(2*image_crop_2_1.size[0]/2)/2+10
top = 0
bottom = height*4+78
imCrop = collage.crop((left, top, right, bottom))
imCrop.save(file_output, "BMP", quality=100)
# collage.show()
#---------------------------------------------------------
os.remove(file_1 + 'Cropped1.bmp')
os.remove(file_1 + 'Cropped2.bmp')
os.remove(file_2 + 'Cropped1.bmp')
os.remove(file_2 + 'Cropped2.bmp')
os.remove(file_3 + 'Cropped1.bmp')
os.remove(path+"merged_image.bmp")
#---------------------------------------------------------
```
